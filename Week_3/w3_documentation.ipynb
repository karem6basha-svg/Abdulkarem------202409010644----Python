# **1. Identify the Components**

##**Inputs:**

Quiz 1 mark

Quiz 2 mark

Quiz 3 mark

Continue choice Y/N


##**Process:**

Calculate the average of three quiz marks.

Check whether the average is 50 or above.

Repeat the process if the user selects Y.

##**Output:**

Average mark

Pass or Fail result

Program Ended message



# **2. Design the algorithm**
##** 2.1 Pseudocode**

START

Set choice = "y"

WHILE choice is equal to "y"

    Enter Quiz 1 mark
    Enter Quiz 2 mark
    Enter Quiz 3 mark

    Calculate average = (Quiz 1 + Quiz 2 + Quiz 3) / 3

    Display average

    IF average >= 50 THEN
        Display "Pass"
    ELSE
        Display "Fail"
    END IF

    Ask user: Continue? Select Y/N

END WHILE

Display "Program Ended"

END

Input quiz marks
